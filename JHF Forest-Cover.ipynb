```python
# load data
train=pd.read_csv("train.csv")
print(len(train))
(train.head(n=3))
```

```python
# Training data
X = train.drop(['Cover_Type', 'Id'], axis=1)
Y = train[['Cover_Type']]
train_data, train_labels = X[:14119], Y[:14119]
dev_data, dev_labels = X[14120:], Y[14120:]

(train_data.head(n=3))
(train_labels.head(n=3))
train_labels.shape
train_labels['Cover_Type'].shape
```

```python
# K-Nearest-Neighbors model where (k = 1)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(train_data, train_labels)
pred = knn.predict(dev_data)
print ("k=1", accuracy_score(dev_labels, pred))

# here must be some code for your training set and test set
tuned_parameters = [{'n_neighbors': np.arange(2)+1,
'weights': ['distance', 'uniform'],
'algorithm': ['ball_tree', 'kd_tree', 'brute']}]
scores = ['precision_weighted', 'recall_weighted']
performances = [];
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()
    clf = GridSearchCV(KNeighborsClassifier(), tuned_parameters, cv=10, scoring=score)
    clf.fit(train_data, train_labels['Cover_Type'])
    print("Best parameters set found on development set:")
    print()
    print(clf.best_estimator_)
    print()
    print("Grid scores on development set:")
    print()
for params, mean_score, scores in clf.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r"
    % (mean_score, scores.std() / 2, params))
    print()
    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = dev_labels, clf.predict(dev_data)
    print(classification_report(y_true, y_pred))

    print()
```

```python
#naive bayes

#instantiate and binarize data
clf = BernoulliNB(binarize=0.5)
#fit model
mod = clf.fit(train_data, train_labels)
#predict on dev
pred = mod.predict(dev_data)
#accuracy score
print ("NB binarized", accuracy_score(dev_labels, pred))

#instantiate and binarize data
clf = GaussianNB()
#fit model
mod = clf.fit(train_data, train_labels)
#predict on dev
pred = mod.predict(dev_data)
#accuracy score
print ("NB Gaussian", accuracy_score(dev_labels, pred))
```
