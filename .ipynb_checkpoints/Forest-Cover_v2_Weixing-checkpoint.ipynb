```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

import re
from sklearn.pipeline import Pipeline
from sklearn.datasets import fetch_mldata
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# SK-learn libraries for evaluation.
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report

# SK-learn libraries for feature extraction from text.
from sklearn.feature_extraction.text import *

# Set the randomizer seed so results are the same each time.
np.random.seed(0)

# Import libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from time import time

# display settings
pd.set_option('display.max_columns',None)
```

```python
# load data
train=pd.read_csv("train.csv")
print('data size:', len(train))
train.head(n=3)
```

```python
# Training data
X = train.drop(['Cover_Type', 'Id'], axis=1)
Y = train[['Cover_Type']]
train_data, train_labels = X[:14119], Y[:14119]
dev_data, dev_labels = X[14120:], Y[14120:]

(train_data.head(n=3))
(train_labels.head(n=3))
#print(train_labels.shape)
print("train label shape", train_labels['Cover_Type'].shape)
```

```python
# check abnormal values and data cleansing
print("Any null values?", train.isnull().values.any())
print(train.describe())
print("\nNo null values, values of all features are in a reasonable range, no abnormal data points")
```

```python
# K-Nearest-Neighbors model where (k = 1)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(train_data, train_labels)
pred = knn.predict(dev_data)
print ("k=1", accuracy_score(dev_labels, pred))

# defining best parameters
tuned_parameters = [{'n_neighbors': np.arange(2)+1,
'weights': ['distance', 'uniform'],
'algorithm': ['ball_tree', 'kd_tree', 'brute']}]
scores = ['precision_weighted', 'recall_weighted']
performances = [];
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()
    clf = GridSearchCV(KNeighborsClassifier(), tuned_parameters, cv=10, scoring=score)
    clf.fit(train_data, train_labels['Cover_Type'])
    print("Best parameters set found on training data:")
    print('\n', clf.best_estimator_)
    print("\nGrid scores on training data:")
for params, mean_score, scores in clf.grid_scores_:
    print("\n%0.3f (+/-%0.03f) for %r"
    % (mean_score, scores.std() / 2, params))
    print("\nDetailed classification report:")
    print("\nThe model is trained on training data")
    print("The scores are computed on the dev data")
    y_true, y_pred = dev_labels, clf.predict(dev_data)
    print('\n', classification_report(y_true, y_pred))
```

```python
#naive bayes

#instantiate and binarize data
clf = BernoulliNB(binarize=0.5)
#fit model
mod = clf.fit(train_data, train_labels)
#predict on dev
pred = mod.predict(dev_data)
#accuracy score
print ("NB binarized", accuracy_score(dev_labels, pred))

#instantiate and binarize data
clf = GaussianNB()
#fit model
mod = clf.fit(train_data, train_labels)
#predict on dev
pred = mod.predict(dev_data)
#accuracy score
print ("NB Gaussian", accuracy_score(dev_labels, pred))
```
